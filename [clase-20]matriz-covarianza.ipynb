```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
# Crear una instancia del StandardScaler para estandarizar características
scaler = StandardScaler()

# Seleccionar las características del conjunto de datos 'iris' a estandarizar
selected_features = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# Aplicar el StandardScaler
scaled = scaler.fit_transform(selected_features)

# Transponer la matriz 'scaled' para invertir la disposición de filas y columnas
scaled.T
```

```python
covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 12},
                 yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                 xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```

### También se puede usar el método de Pandas: dataframe.corr()

```python
corr_matrix = iris.select_dtypes(include='number').corr() 
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot = True)
```
